```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('PS18_Dataset_18.csv')

print("---Original Data Overview---")
print(df.head())

clean_df = df.dropna()
print("\n---Cleaned Data Overview---")
print(clean_df.info())


average_A = np.mean(df['A'])
highest_A = np.max(df['A'])
print("\nAverage of Column A:", average_A)
print("Highest number in Column A:", highest_A)


average_B = np.mean(df['B'])
highest_B = np.max(df['B'])
print("\nAverage of Column B:", average_B)
print("Highest number in Column B:", highest_B)

average_C = np.mean(df['C'])
highest_C = np.max(df['C'])
print("\nAverage of Column C:", average_C)
print("Highest number in Column C:", highest_C)


average_D = np.mean(df['D'])
highest_D = np.max(df['D'])
print("\nAverage of Column D:", average_D)
print("Highest number in Column D:", highest_D)

columns = ['A', 'B', 'C', 'D']
colors=['blue','green','red','yellow']
averages = [np.mean(df['A']), np.mean(df['B']), np.mean(df['C']), np.mean(df['D'])]

plt.figure(figsize=(7, 4))
plt.bar(columns, averages,color=colors)
plt.title('\nAverage Values of All Columns')
plt.show()


plt.scatter(df['A'], df['B'], c=df['C'], cmap='viridis')
plt.colorbar(label='Column C value')
plt.xlabel('A')
plt.ylabel('B')
plt.title('Scatter: A vs B colored by C')
```
